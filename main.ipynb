```python
from datetime import datetime
import pandas as pd
import talib
```

```python
df_rates = pd.read_pickle('./data/df_rates_3H.pkl')
df_rates
```

```python
def green_bar(row):
    if (row['rsi'] > 50) and (row['stoch_k'] > 50) and (row['macd_hist'] > 0):
        return int(1)
    else:
        return int(0)
    
def red_bar(row):
    if (row['rsi'] < 50) and (row['stoch_k'] < 50) and (row['macd_hist'] < 0):
        return int(-1)
    else:
        return int(0)
    
def black_bar(row):
    if (row['green_bar'] == 0) and (row['red_bar'] == 0):
        return int(1)
    else:
        return int(0)
```

```python
df_rates['rsi'] = talib.RSI(df_rates['close'], timeperiod=7)
df_rates['stoch_k'], df_rates['stoch_d'] = talib.STOCH(df_rates['high'], df_rates['low'], df_rates['close'],
                                                       fastk_period=14, slowk_period=3, slowd_period=3)
df_rates['macd'], df_rates['macd_signal'], df_rates['macd_hist'] = talib.MACD(df_rates['close'],
                                                                fastperiod=12, slowperiod=26, signalperiod=9)

# df_rates['time_jst'] = df_rates['time'].dt.k

df_rates['green_bar'] = df_rates.apply(green_bar, axis=1)
df_rates['red_bar'] = df_rates.apply(red_bar, axis=1)
df_rates['black_bar'] = df_rates.apply(black_bar, axis=1)


df_rates
```
